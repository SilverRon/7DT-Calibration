```python
# Plot presetting
import matplotlib.pyplot as plt
import matplotlib as mpl
#
mpl.rcParams["axes.titlesize"] = 14
mpl.rcParams["axes.labelsize"] = 20
plt.rcParams['savefig.dpi'] = 500
plt.rc('font', family='serif')

# %%
plt.style.use('dark_background') # Dark 모드 스타일 적용
```

```python
path_table = '../output/*/result.csv'
intablelist = sorted(glob.glob(path_table))
print(intablelist)
```

```python
bvlist = [-0.30, +0.55, 0.06]
vlist = [11.83, 11.51, 12.21]
```

```python
intbl_list = [Table.read(intable) for intable in intablelist]
```

```python
# for filte in ['u', 'g', 'r', 'i', 'z']:
for filte in ['g', 'r', 'i', 'z']:
	zperrlist = []
	for nn in range(len(intbl_list)):
		intbl = intbl_list[nn]
		zperr = intbl['zperr_gaia'][intbl['filter']==filte].item()
		zperrlist.append(zperr)

	#
	# Sort 'bvlist' and get the indices that would sort it
	sorted_indices = sorted(range(len(bvlist)), key=lambda k: bvlist[k])

	# Sort 'bvlist' and 'zperrlist' based on the sorted indices
	sorted_bvlist = [bvlist[i] for i in sorted_indices]
	sorted_zperrlist = [zperrlist[i] for i in sorted_indices]

	plt.plot(sorted_bvlist, sorted_zperrlist, 'o-', label=filte)

plt.ylim([0, 0.1])
# plt.legend()
plt.legend(loc='upper center', ncol=4, fontsize=12)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('B-V')
plt.ylabel(r'$\rm ZP_{Gaia}$ Error')
plt.tight_layout()
plt.savefig("../output/result.brd.zp.png", transparent=True)
```
