```python
def calculate_magnitude_error(flux, flux_error):
    snr = flux / flux_error
    magnitude_error = 2.5*np.log10(1+(1/snr))
    
    return magnitude_error
```

```python
intbl = Table.read(f'../output/{tname}.pre.csv')
indf = intbl.to_pandas()
```

```python
import pandas as pd

# Create DataFrames for Table A and Table B

# Merge the tables based on 'source_id'
merged_df = pd.merge(synthetic_photometry, indf, on='source_id')
```

```python
flux = merged_df['USER_7DT_Edmund_flux_400_25']
flux_error = merged_df['USER_7DT_Edmund_flux_error_400_25']
calculate_magnitude_error(flux, flux_error)
```

```python
magprefix = 'USER_7DT_Edmund_mag_'
magerrprefix = 'USER_7DT_Edmund_magerr_'
fluxprefix = 'USER_7DT_Edmund_flux_'
fluxerrprefix = 'USER_7DT_Edmund_flux_error_'


for key in merged_df.keys():
	if magprefix in key:
		#	Filter
		filte = key.replace(magprefix, '')
		#	Flux & Flux Error
		fluxkey = key.replace(magprefix, fluxprefix)
		fluxerrkey = key.replace(magprefix, fluxerrprefix)
		#	Magnitude Error
		newkey = key.replace(magprefix, magerrprefix)
		_flux = merged_df[fluxkey]
		_flux_error = merged_df[fluxerrkey]
		magerr = calculate_magnitude_error(_flux, _flux_error)
		merged_df[newkey] = magerr
```

```python
merged_df
```

```python
key_to_rename_dict = {}

prefix_mag_7dt = 'USER_7DT_Edmund_mag_'
prefix_magerr_7dt = 'USER_7DT_Edmund_magerr_'
prefix_mag_sm = 'skymapper_mag_'

for key in merged_df.keys():
	for prefix in [prefix_mag_7dt, prefix_magerr_7dt]:
		if (prefix in key):
			if ('_50' in key[-3:]):
				newkey = key.replace(prefix, 'm').replace('_50', 'w')
			elif ('_25' in key[-3:]):
				newkey = key.replace(prefix, 'm').split('_')[-2]
			else:
				newkey = key.replace(prefix, 'm')
			if 'err' in prefix:
				newkey = f"{newkey}err"
			key_to_rename_dict[key] = newkey

	if (prefix_mag_sm in key):
		newkey = key.replace(prefix_mag_sm, '')
		key_to_rename_dict[key] = newkey

key_to_rename_dict
```

```python
filterlist = ['u', 'u2', 'v', 'g', 'r', 'i', 'z']
for filte in filterlist:
	merged_df[f"{filte}err"] = 0.0
```

```python
merged_df.rename(columns=key_to_rename_dict, inplace=True)
merged_df
```

```python
merged_df.to_csv(f'../output/gaia-{tname}.cat')
```
