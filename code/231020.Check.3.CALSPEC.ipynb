```python
import os, glob, sys
import matplotlib.pyplot as plt
from astropy import units as u
from astropy.io import fits
import numpy as np
from astropy.table import Table, vstack, hstack
from astropy import units as u
from astropy.coordinates import SkyCoord
import warnings
warnings.filterwarnings("ignore")

# Plot presetting
import matplotlib.pyplot as plt
import matplotlib as mpl
#
mpl.rcParams["axes.titlesize"] = 14
mpl.rcParams["axes.labelsize"] = 20
plt.rcParams['savefig.dpi'] = 500
plt.rc('font', family='serif')

# %%
plt.style.use('dark_background') # Dark 모드 스타일 적용
```

```python
intbl1 = Table.read('../data/Xshooter/fFeige110.dat', format='ascii')
intbl2 = Table.read('../data/Xshooter/fLTT7987.dat', format='ascii')
intbl3 = Table.read('../data/Xshooter/fltt1020.dat', format='ascii')
```

```python
plt.plot(intbl1['col1'], intbl1['col2']/np.max(intbl1['col2']), label='Feige 110')
plt.plot(intbl2['col1'], intbl2['col2']/np.max(intbl2['col2']), label='LTT 7987')
plt.plot(intbl3['col1'], intbl3['col2']/np.max(intbl3['col2']), label='LTT 1020')
plt.legend(fontsize=14)
plt.xlim([3750, 9000])
plt.xlabel(r'Wavelength [$\rm \AA$]')
plt.ylabel(r'Normalized Flux')
plt.tight_layout()
```
